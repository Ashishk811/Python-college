### IP Address version (IPv4 or IPv6)

```python
import ipaddress
def check_ip_version(ip_add_str):
    try:
        ip_obj = ipaddress.ip_address(ip_add_str)
        if ip_obj.version == 4:
            return " IPv4"
        elif ip_obj.version == 4:
            return " IPv6"
    except ValueError:
        return "Invalid IP Address"

print("172.28.0.12 is: ", check_ip_version('172.28.0.12'))
print("2001:db8::1 is: ", check_ip_version('2001:db8::1'))
print("invalid_ip is: ", check_ip_version('invalid_ip'))
```
