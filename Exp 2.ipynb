# Experiment 2: String Operations

### Sequence Type

```python
my_str =  "Python"
print(my_str[0])
print(my_str[1:4])
```

### Basic Operations

```python
# Concatenation
s1= "Hello"
s2= " world!"
str_sum= s1 + s2
print(str_sum)

# Repeatation
print("Hello " *5 )

# Length
print(len("No Hello"))
```

### Case Conversion

```python
str1= "hEllO pYthoN"
print(str1.upper())
print(str1.lower())
print(str1.capitalize())
print(str1.title())
```

### Searching & Replacing

```python
str1= "hEllo pYthoNNN"
print(str1.find("o"))
print(str1.rfind("o"))
print(str1.replace("hEllo", "Hi" ))
print(str1.count("N"))
```

### Splitting & Joining

```python
str1= "This is a good mango"
str_split= str1.split()
print(str_split)
str_join= "--".join(str_split)
print(str_join)
```

### Stripping White-space

```python
str1= " Hello "
print(str1)
print(str1.strip())
print(str1.rstrip())
print(str1.lstrip())
```

### Checking Content

```python
s1= "Hello"
n1= "1245"
an1= "Hello512"
sp1= " "
print(s1.isalpha())
print(an1.isalpha())

print(n1.isdigit())

print(an1.isalnum())

print(sp1.isspace())
print(an1.isspace())
```

### Examples

```python
msg = " Wednesday Addams! "

# Accessing Charcters
print(msg[3])

# Slicing
print(msg[3:10])

# Stripping white-space
print(msg.strip())

# Upper-case
print(msg.upper())

# Replacing
print(msg.replace("Wednesday","Friday"))
```

### Looping through a string

```python
for x in "banana":
    print(x)
```

### Check the sub-string

```python
txt= "This is a good mango"
print("good" in txt)

if "is" in txt:
    print("Persent")

if "am" not in txt:
    print("Not Present")
```
