### Starting a Thread

```python
from threading import Thread
def info (bar):
    print("hello {}".format(bar))  # hello + {input}
    return 'info'

thread = Thread(target=info, args=('world!',))
thread.start()
return_value=thread.join()
print(return_value)
```

```python
from threading import Thread

def cube (num):
    print("Cube: ",num*num*num)

def square(num):
    print(f"Square: {num*num}")
    
if __name__=='__main__':
    t1= Thread(target=cube,args=(6,))
    t2= Thread(target=square,args=(5,))
    
    t1.start()
    t2.start()
    
    # Join the thread, blocking the main thread until my_thread completes
    t1.join()
    t2.join()
```

```python
import threading
import time

def worker_fun(name):
    """
    A function to be executed by a thread.
    Simulates a task taking some time.
    """
    print(f"Thread {name}: Starting--->")
    time.sleep(2) #time consuming operations (Asynchronomus- setTimeout() function in JS)
    print(f"Thread {name}: finishing--->")

if __name__=='__main__':
    print("Main thread: creating and starting thread")
    t1= threading.Thread(target=worker_fun,args=("One",))
    t2= threading.Thread(target=worker_fun,args=("Two",))
    
    t1.start()
    t2.start()    
    
    print("Main thread: Doing other work while thread runs")
    time.sleep(1) #simulate main thraed doing other work
    
    # wait for main threads to complete before the main thread exits
    t1.join()
    t2.join()
    
    print("Main thread: Thread finished")
```

```python
nums=[1,2,3,4,5]

def cube (num):
    time.sleep(0.5)
    print(f"Cube of {num}: {num*num*num}")
    

def square(nums):
    time.sleep(0.5)
    print(f"Square of {num}: {num*num}")
    

if __name__=='__main__':
    start_time = time.perf_counter()
    
    print("-----WITHOUT THREADING-----")
    for i in nums:
        cube(i)
    for i in nums:
        square(i)
    
    end_time = time.perf_counter()
    total_execution_time = end_time - start_time
    print(f"\nAll threads completed for WITHOUT threading: {total_execution_time} seconds.")
    
    print("\n-----WITH THREADING----")
    
    start_time = time.perf_counter()
    for i in nums:
        t1= Thread(target=cube,args=(i,))
        t2= Thread(target=square,args=(i,))

        t1.start()
        t2.start()
        

        t1.join()
        t2.join()
        
    end_time = time.perf_counter()
    total_execution_time = end_time - start_time
    print(f"\nAll threads completed for WITH threading: {total_execution_time} seconds.")
```
